```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
base_url = 'http://www.weather.go.kr/weather/climate/past_cal.jsp?stn=108&yy=2017&mm={}&obs=1&x=35&y=10'
```

```python
for n in range (12) :
    url = base_url.format(n+1)
    webpage = urlopen(url)
    source = BeautifulSoup(webpage,'html5lib')
    my_titles = source.find_all('td',{'class':'align_left'})
    
    review_list = []
    for title in my_titles:
        review_list.append(title.get_text().strip())
        print(title.get_text().strip())
```

```python
file = open('data1.txt', 'w', encoding = 'utf-8')
 
for title in review_list:
    file.write(title + '\n')
    
file.close()
```
